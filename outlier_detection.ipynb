```python
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor

# Step 1: Load the dataset
df = pd.read_csv('olympics.csv')  # Replace with your actual path if needed
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df = df.drop(index=0).reset_index(drop=True)
```

```python
df.columns = ['Country', 'Summer_Games', 'Summer_Gold', 'Summer_Silver', 'Summer_Bronze', 'Summer_Total',
              'Winter_Games', 'Winter_Gold', 'Winter_Silver', 'Winter_Bronze', 'Winter_Total',
              'Combined_Games', 'Combined_Gold', 'Combined_Silver', 'Combined_Bronze', 'Combined_Total']
```

```python
df.head()
```

```python

# Step 2: Select numeric columns and drop rows with NaNs
numeric_df = df.select_dtypes(include=['int64', 'float64']).dropna()
```

```python
# Check if we actually have any numeric data
if numeric_df.shape[1] == 0:
    raise ValueError("No numeric columns found for outlier detection.")
```

```python
# Step 3: Apply Local Outlier Factor (LOF)
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # 5% as potential outliers
outliers = lof.fit_predict(numeric_df)
```

```python
# Step 4: Filter out outliers (-1 = outlier, 1 = inlier)
df_clean = df[outliers == 1]
```

```python
# Step 5: Print data counts before and after
print("Original data count:", df.shape[0])
print("Data count after outlier removal:", df_clean.shape[0])
```
