```python
import pandas as pd
import scipy.stats as stats


# Assuming your file name is 'diabetes.csv'
df = pd.read_csv('diabetes.csv')

# Perform univariate analysis
univariate_analysis = {
    'Mean': df.mean(),
    'Median': df.median(),
    'Variance': df.var(),
    'Standard Deviation': df.std(),
    'Skewness': df.skew(),
    'Kurtosis': df.kurtosis()
}

# Convert the dictionary to a DataFrame for better display
univariate_analysis_df = pd.DataFrame(univariate_analysis)

# Display the results
print("Univariate Analysis Results:")
display(univariate_analysis_df)
```
