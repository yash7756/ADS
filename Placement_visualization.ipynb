```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
sns.histplot(df['cgpa'], kde=True, color='skyblue')
plt.title('Distribution of CGPA')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['placement_exam_marks'], kde=True, color='lightgreen')
plt.title('Distribution of Placement Exam Marks')
plt.xlabel('Exam Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x='placed', data=df, palette='Set2')
plt.title('Placement Status Count')
plt.xlabel('Placed (0 = No, 1 = Yes)')
plt.ylabel('Number of Students')
plt.show()
```

```python
sns.scatterplot(x='cgpa', y='placement_exam_marks', hue='placed', data=df, palette='Set1')
plt.title('CGPA vs Placement Exam Marks')
plt.xlabel('CGPA')
plt.ylabel('Exam Marks')
plt.legend(title='Placed')
plt.show()
```

```python
sns.boxplot(x='placed', y='cgpa', data=df, palette='pastel')
plt.title('CGPA Distribution by Placement Status')
plt.xlabel('Placed')
plt.ylabel('CGPA')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, hue='placed', palette='husl')
plt.suptitle('Pairplot of Placement Dataset', y=1.02)
plt.show()
```
