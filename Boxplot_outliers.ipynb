```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('placement.csv')  # Make sure 'placement.csv' is uploaded to your Colab session

# Visualizing box plots for CGPA and Placement Exam Marks
plt.figure(figsize=(12, 6))

# Box plot for CGPA
plt.subplot(1, 2, 1)
sns.boxplot(y=df['cgpa'])
plt.title('Box Plot of CGPA')

# Box plot for Placement Exam Marks
plt.subplot(1, 2, 2)
sns.boxplot(y=df['placement_exam_marks'])
plt.title('Box Plot of Placement Exam Marks')

plt.tight_layout()
plt.show()

# Identifying outliers using IQR
def find_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Get outliers for both columns
cgpa_outliers = find_outliers('cgpa')
exam_marks_outliers = find_outliers('placement_exam_marks')

print("Outliers in CGPA:")
print(cgpa_outliers)

print("\nOutliers in Placement Exam Marks:")
print(exam_marks_outliers)
```
